```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Graph Setup
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 129Channels Data
df = pd.read_csv('129Channels.csv')
xdata = df['X']
ydata = df['Y']
zdata = df['Z']
ind = df['Unnamed: 0']

# plot all the points
ax.scatter3D(xdata, ydata, zdata, color='b')

# label the points
for i in range(len(xdata)):
    ax.text(xdata[i], ydata[i], zdata[i], ind[i], size=7)
```

```python
# Graph Setup
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

df = pd.read_csv('129Channels.csv')
ds = pd.read_csv('PeripheralChannels.csv')

# 129Channels Data
xdata = df['X']
ydata = df['Y']
zdata = df['Z']
ind = df['Unnamed: 0']

# Peripheral Channels Data
x2data = ds['X']
y2data = ds['Y']
z2data = ds['Z']
ind2 = ds['Unnamed: 0']

s2 = set(ind2.values.tolist())
nChannel = False
pChannel = False

# For every point, if it is a Peripheral Channel plot it red, else plot blue
for i in range(len(xdata)):
    if ind[i] in s2:
        ax.scatter(xdata[i], ydata[i], zdata[i], color='r', label="Peripheral Channel" if pChannel==False else "")
        ax.text(xdata[i], ydata[i], zdata[i], ind[i], size=7)
        pChannel = True
    else: 
        ax.scatter(xdata[i], ydata[i], zdata[i], color='b', label="Normal Channel" if nChannel==False else "")
        ax.text(xdata[i], ydata[i], zdata[i], ind[i], size=7)
        nChannel = True
        
# Legend
plt.legend(loc="upper left")
```
